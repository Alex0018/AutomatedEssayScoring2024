```python
def get_num_words(text):
    return text.count(' ')

def ttr(text):
    # Tokenize text
    tokens = nltk.word_tokenize(text.lower())

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

    # Calculate vocabulary size
    vocab_size = len(set(filtered_tokens))

    # Calculate Type-Token Ratio (TTR)
    total_tokens = len(filtered_tokens)
    ttr = vocab_size / total_tokens if total_tokens > 0 else 0
    return ttr


stat_dict = {'symbol_length':  len,
             'num_words':      get_num_words,
             'flesch_kincaid': textstat.flesch_kincaid_grade,
             'gunning_fog':    textstat.gunning_fog,
             'smog_index':     textstat.smog_index,
             'ttr': ttr}

df_train = read_train()
for feature, mapper in stat_dict.items():
    df_train[feature] = df_train['full_text'].map(mapper)
```

# LGB baseline with error count features

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import cohen_kappa_score
from lightgbm import LGBMClassifier
import lightgbm

lgb = LGBMClassifier(random_state=42, n_estimators=1000)

drop_features = ['essay_id', 'label', 'fold']

for fold in range(df_train['fold'].nunique()):
    X_train = df_train.query(f'fold!={fold}')
    Y_train = X_train['label']
    X_train = X_train.drop(drop_features, axis=1)

    X_val = df_train.query(f'fold=={fold}')
    Y_val = X_val['label']
    X_val = X_val.drop(drop_features, axis=1)

    lgb.fit(X_train, Y_train, 
            eval_set=[(X_val, Y_val)], 
            callbacks=[lightgbm.early_stopping(stopping_rounds=50, verbose=0), lightgbm.log_evaluation(0)],
            )
    preds = lgb.predict(X_val)

    score = cohen_kappa_score(Y_val, preds, weights='quadratic')

    print(f'Fold {fold}:    {score:.3f}      {lgb.best_iteration_} estimators')
```

# LGB baseline with error frequency features

```python
import warnings 
warnings.filterwarnings('ignore')

def get_features(df):
    drop_features = ['essay_id', 'label', 'fold']
    for feature in df.columns[9:]:
        df[feature] /= df['num_words']
    return df.drop(drop_features, axis=1)


for fold in range(df_train['fold'].nunique()):
    X_train = df_train.query(f'fold!={fold}')
    Y_train = X_train['label']
    X_train = get_features(X_train)

    X_val = df_train.query(f'fold=={fold}')
    Y_val = X_val['label']
    X_val = get_features(X_val)

    lgb.fit(X_train, Y_train, 
            eval_set=[(X_val, Y_val)], 
            callbacks=[lightgbm.early_stopping(stopping_rounds=50, verbose=0), lightgbm.log_evaluation(0)],
            )
    preds = lgb.predict(X_val)

    score = cohen_kappa_score(Y_val, preds, weights='quadratic')

    print(f'Fold {fold}:    {score:.3f}      {lgb.best_iteration_} estimators')
```
